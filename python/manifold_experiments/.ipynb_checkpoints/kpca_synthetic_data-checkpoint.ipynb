```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap, LocallyLinearEmbedding, MDS, SpectralEmbedding
from sklearn.decomposition import FastICA, NMF
from sklearn.random_projection import SparseRandomProjection
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

def plot_dimensionality_reduction(X, y, figsize=(12, 8)):
    """
    Plot data using various non-linear dimensionality reduction methods.
    
    Parameters:
    X : array-like, shape (n_samples, n_features)
        The input data
    y : array-like, shape (n_samples,)
        The labels for coloring points
    figsize : tuple, optional
        Figure size for the plot
    """
    
    # Standardize the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Define the dimensionality reduction methods
    methods = {
        'Isomap': Isomap(n_components=2, n_neighbors=10),
        'MDS': MDS(n_components=2, random_state=42),
        'FastICA': FastICA(n_components=2, random_state=42),
        'NMF': NMF(n_components=2, random_state=42),
        'Sparse Random Projection': SparseRandomProjection(n_components=2, random_state=42)
    }
    
    # Create subplots
    fig, axes = plt.subplots(2, 3, figsize=figsize)
    axes = axes.ravel()
    
    # Get unique labels for consistent coloring
    unique_labels = np.unique(y)
    colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))
    
    for i, (name, method) in enumerate(methods.items()):
        try:
            print(f"Computing {name}...")
            
            # Handle NMF which requires non-negative input
            if name == 'NMF':
                # Shift data to be non-negative
                X_input = X_scaled - X_scaled.min() + 1e-6
            else:
                X_input = X_scaled
            
            # Fit and transform the data
            X_transformed = method.fit_transform(X_input)
            
            # Plot the results
            ax = axes[i]
            for j, label in enumerate(unique_labels):
                mask = y == label
                ax.scatter(X_transformed[mask, 0], X_transformed[mask, 1], 
                          c=[colors[j]], label=f'Class {label}', alpha=0.7, s=30)
            
            ax.set_title(f'{name}', fontsize=12, fontweight='bold')
            ax.set_xlabel('Component 1')
            ax.set_ylabel('Component 2')
            ax.grid(True, alpha=0.3)
            ax.legend(fontsize=8)
            
        except Exception as e:
            print(f"Error with {name}: {str(e)}")
            axes[i].text(0.5, 0.5, f'{name}\nFailed', 
                        horizontalalignment='center', verticalalignment='center',
                        transform=axes[i].transAxes, fontsize=12)
            axes[i].set_title(f'{name} (Failed)', fontsize=12)
    
    plt.tight_layout()
    #plt.suptitle('Non-linear Dimensionality Reduction Methods', 
    #             fontsize=16, fontweight='bold', y=1.02)
    plt.show()

def compare_methods_metrics(X, y):
    """
    Compare different methods and print some basic metrics.
    """
    from sklearn.metrics import silhouette_score
    from sklearn.neighbors import NearestNeighbors
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    methods = {
        'Isomap': Isomap(n_components=2, n_neighbors=10),
        'LLE': LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42),
        'MDS': MDS(n_components=2, random_state=42),
        'Spectral Embedding': SpectralEmbedding(n_components=2, random_state=42)
    }
    
    print("\nMethod Comparison (Silhouette Scores):")
    print("-" * 40)
    
    for name, method in methods.items():
        try:
            X_transformed = method.fit_transform(X_scaled)
            score = silhouette_score(X_transformed, y)
            print(f"{name:20}: {score:.3f}")
        except Exception as e:
            print(f"{name:20}: Failed ({str(e)[:30]}...)")
```

```python
plot_dimensionality_reduction(X, y)
compare_methods_metrics(X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA, KernelPCA
from sklearn.manifold import Isomap, LocallyLinearEmbedding
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
from sklearn.neighbors import NearestNeighbors
from scipy.spatial.distance import pdist, squareform
from scipy.stats import pearsonr, spearmanr
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)

class DatasetGenerator:
    """Generate synthetic datasets with varying degrees of nonlinearity"""
    
    @staticmethod
    def generate_linear_dataset(n_samples=2500, intrinsic_dim=35, ambient_dim=250):
        """Generate linear dataset using random projection"""
        # Create intrinsic data
        X_intrinsic = np.random.randn(n_samples, intrinsic_dim)
        
        # Random projection matrix
        projection_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.5
        
        # Project to high dimension
        X_ambient = X_intrinsic @ projection_matrix
        
        # Add small amount of noise
        noise = np.random.randn(n_samples, ambient_dim) * 0.05
        X_ambient += noise
        
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_swiss_roll_high_dim(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate high-dimensional Swiss roll manifold"""
        # Create base Swiss roll in 3D
        t = np.random.uniform(1.5 * np.pi, 4.5 * np.pi, n_samples)
        x = t * np.cos(t)
        y = np.random.uniform(0, 20, n_samples)
        z = t * np.sin(t)
        
        base_manifold = np.column_stack([x, y, z])
        
        # Extend to higher intrinsic dimensions with coupled nonlinear relationships
        X_intrinsic = np.zeros((n_samples, intrinsic_dim))
        X_intrinsic[:, :3] = base_manifold
        
        for i in range(3, intrinsic_dim):
            if i % 3 == 0:
                X_intrinsic[:, i] = np.sin(t) * np.cos(y/5) + np.random.randn(n_samples) * 0.2
            elif i % 3 == 1:
                X_intrinsic[:, i] = np.cos(t) * np.sin(y/5) + np.random.randn(n_samples) * 0.2
            else:
                X_intrinsic[:, i] = np.tanh(x/10) * np.log(1 + y/5) + np.random.randn(n_samples) * 0.2
        
        # Embed in high-dimensional space using nonlinear mapping
        embedding_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.3
        
        # Apply nonlinear transformations
        X_ambient = np.zeros((n_samples, ambient_dim))
        for i in range(ambient_dim):
            linear_comb = X_intrinsic @ embedding_matrix[:, i]
            if i % 4 == 0:
                X_ambient[:, i] = np.tanh(linear_comb)
            elif i % 4 == 1:
                X_ambient[:, i] = np.sin(linear_comb)
            elif i % 4 == 2:
                X_ambient[:, i] = linear_comb * np.exp(-linear_comb**2/10)
            else:
                X_ambient[:, i] = np.sign(linear_comb) * np.sqrt(np.abs(linear_comb))
        
        # Add noise
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_sphere_manifold(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate high-dimensional spherical manifold with complex structure"""
        # Start with hypersphere sampling
        X_intrinsic = np.random.randn(n_samples, intrinsic_dim)
        norms = np.linalg.norm(X_intrinsic, axis=1, keepdims=True)
        X_intrinsic = X_intrinsic / norms
        
        # Add complexity with coupled dimensions
        for i in range(intrinsic_dim//2):
            j = (i + intrinsic_dim//4) % intrinsic_dim
            k = (i + intrinsic_dim//2) % intrinsic_dim
            X_intrinsic[:, i] *= np.cos(X_intrinsic[:, j] * np.pi)
            X_intrinsic[:, i] += 0.3 * np.sin(X_intrinsic[:, k] * 2 * np.pi)
        
        # Renormalize
        norms = np.linalg.norm(X_intrinsic, axis=1, keepdims=True)
        X_intrinsic = X_intrinsic / norms
        
        # Complex nonlinear embedding
        embedding_weights = np.random.randn(intrinsic_dim, ambient_dim) * 0.4
        
        X_ambient = np.zeros((n_samples, ambient_dim))
        for i in range(ambient_dim):
            weights = embedding_weights[:, i]
            projection = X_intrinsic @ weights
            
            if i % 5 == 0:
                X_ambient[:, i] = projection**3
            elif i % 5 == 1:
                X_ambient[:, i] = np.exp(projection) * np.cos(projection * 3)
            elif i % 5 == 2:
                X_ambient[:, i] = np.sinh(projection) / (1 + projection**2)
            elif i % 5 == 3:
                X_ambient[:, i] = projection * np.sin(projection**2)
            else:
                X_ambient[:, i] = np.arctan(projection * 5) * np.cos(projection)
        
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_hypercube_corners(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate data concentrated near hypercube corners - simulates discrete activations"""
        # Create sparse, clustered structure
        n_clusters = max(8, intrinsic_dim // 3)
        cluster_centers = np.random.choice([-1, 1], size=(n_clusters, intrinsic_dim))
        
        # Assign samples to clusters
        cluster_assignments = np.random.choice(n_clusters, size=n_samples)
        
        X_intrinsic = np.zeros((n_samples, intrinsic_dim))
        for i in range(n_samples):
            cluster_id = cluster_assignments[i]
            # Add noise around cluster center
            X_intrinsic[i] = cluster_centers[cluster_id] + np.random.randn(intrinsic_dim) * 0.3
            
            # Apply activation-like nonlinearity
            X_intrinsic[i] = np.tanh(X_intrinsic[i] * 2)
        
        # Complex embedding with interaction terms
        embedding_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.4
        interaction_matrix = np.random.randn(intrinsic_dim, intrinsic_dim, ambient_dim // 4) * 0.2
        
        X_ambient = np.zeros((n_samples, ambient_dim))
        
        # Linear terms
        X_ambient[:, :ambient_dim//2] = X_intrinsic @ embedding_matrix[:, :ambient_dim//2]
        
        # Quadratic interaction terms
        for i in range(ambient_dim//4):
            quad_term = np.sum(X_intrinsic[:, :, None] * X_intrinsic[:, None, :] * 
                             interaction_matrix[:, :, i][None, :, :], axis=(1, 2))
            X_ambient[:, ambient_dim//2 + i] = quad_term
        
        # Higher-order nonlinear terms
        for i in range(ambient_dim//4, ambient_dim//2):
            idx = i - ambient_dim//4
            if idx < intrinsic_dim:
                X_ambient[:, ambient_dim//2 + ambient_dim//4 + idx] = (
                    np.sin(X_intrinsic[:, idx] * np.pi) * 
                    np.cos(X_intrinsic[:, (idx + 1) % intrinsic_dim] * np.pi)
                )
        
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_torus_manifold(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate high-dimensional torus manifold with multiple frequency components"""
        # Create base torus parameters
        n_tori = intrinsic_dim // 2
        
        X_intrinsic = np.zeros((n_samples, intrinsic_dim))
        
        # Generate torus coordinates
        for i in range(n_tori):
            theta = np.random.uniform(0, 2*np.pi, n_samples)
            phi = np.random.uniform(0, 2*np.pi, n_samples)
            
            # Major and minor radii with variation
            R = 2 + 0.5 * np.sin(i * np.pi / n_tori)  # Major radius
            r = 0.8 + 0.3 * np.cos(i * np.pi / n_tori)  # Minor radius
            
            X_intrinsic[:, 2*i] = (R + r * np.cos(phi)) * np.cos(theta)
            if 2*i + 1 < intrinsic_dim:
                X_intrinsic[:, 2*i + 1] = (R + r * np.cos(phi)) * np.sin(theta)
        
        # Fill remaining dimensions with coupled oscillations
        for i in range(2*n_tori, intrinsic_dim):
            base_idx = i % (2*n_tori)
            X_intrinsic[:, i] = np.sin(X_intrinsic[:, base_idx] * (i + 1)) * np.cos(X_intrinsic[:, (base_idx + 1) % (2*n_tori)] * (i + 1))
        
        # Complex embedding with frequency modulation
        embedding_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.3
        
        X_ambient = np.zeros((n_samples, ambient_dim))
        for i in range(ambient_dim):
            linear_comb = X_intrinsic @ embedding_matrix[:, i]
            
            # Apply different nonlinearities based on frequency content
            freq = (i % 7) + 1
            if i % 7 == 0:
                X_ambient[:, i] = np.sin(linear_comb * freq) * np.exp(-linear_comb**2/20)
            elif i % 7 == 1:
                X_ambient[:, i] = np.cos(linear_comb * freq) * np.tanh(linear_comb)
            elif i % 7 == 2:
                X_ambient[:, i] = linear_comb * np.sin(linear_comb**2 * freq/10)
            elif i % 7 == 3:
                X_ambient[:, i] = np.sinh(linear_comb/2) * np.cos(linear_comb * freq)
            elif i % 7 == 4:
                X_ambient[:, i] = np.arctan(linear_comb * 3) * np.sin(freq * np.pi * linear_comb)
            elif i % 7 == 5:
                X_ambient[:, i] = (linear_comb / (1 + linear_comb**2)) * np.cos(freq * linear_comb)
            else:
                X_ambient[:, i] = np.sign(linear_comb) * np.sqrt(np.abs(linear_comb)) * np.sin(freq * linear_comb)
        
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_mixture_of_manifolds(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate mixture of different manifold types - simulates heterogeneous activations"""
        # Divide samples among different manifold types
        n_per_manifold = n_samples // 4
        
        X_ambient_total = []
        X_intrinsic_total = []
        
        # Manifold 1: Sphere-like
        X1_amb, X1_int = DatasetGenerator.generate_sphere_manifold(
            n_per_manifold, intrinsic_dim//4, ambient_dim, noise
        )
        # Pad intrinsic dimensions
        X1_int_padded = np.zeros((n_per_manifold, intrinsic_dim))
        X1_int_padded[:, :intrinsic_dim//4] = X1_int
        X_ambient_total.append(X1_amb)
        X_intrinsic_total.append(X1_int_padded)
        
        # Manifold 2: Torus-like
        X2_amb, X2_int = DatasetGenerator.generate_torus_manifold(
            n_per_manifold, intrinsic_dim//3, ambient_dim, noise
        )
        X2_int_padded = np.zeros((n_per_manifold, intrinsic_dim))
        X2_int_padded[:, intrinsic_dim//4:intrinsic_dim//4 + intrinsic_dim//3] = X2_int
        X_ambient_total.append(X2_amb)
        X_intrinsic_total.append(X2_int_padded)
        
        # Manifold 3: Hypercube corners
        X3_amb, X3_int = DatasetGenerator.generate_hypercube_corners(
            n_per_manifold, intrinsic_dim//3, ambient_dim, noise
        )
        X3_int_padded = np.zeros((n_per_manifold, intrinsic_dim))
        start_idx = intrinsic_dim//4 + intrinsic_dim//3
        X3_int_padded[:, start_idx:start_idx + intrinsic_dim//3] = X3_int
        X_ambient_total.append(X3_amb)
        X_intrinsic_total.append(X3_int_padded)
        
        # Manifold 4: Linear subspace with nonlinear distortion
        remaining_samples = n_samples - 3 * n_per_manifold
        remaining_intrinsic_dim = intrinsic_dim - (intrinsic_dim//4 + 2*(intrinsic_dim//3))
        
        # Generate linear subspace
        X4_base = np.random.randn(remaining_samples, remaining_intrinsic_dim)
        # Apply nonlinear distortion
        for i in range(remaining_intrinsic_dim):
            X4_base[:, i] = np.tanh(X4_base[:, i]) + 0.3 * np.sin(X4_base[:, i] * 3)
        
        # Embed to ambient space
        embedding_4 = np.random.randn(remaining_intrinsic_dim, ambient_dim) * 0.4
        X4_amb = X4_base @ embedding_4
        X4_amb += np.random.randn(remaining_samples, ambient_dim) * noise
        
        X4_int_padded = np.zeros((remaining_samples, intrinsic_dim))
        X4_int_padded[:, -remaining_intrinsic_dim:] = X4_base
        X_ambient_total.append(X4_amb)
        X_intrinsic_total.append(X4_int_padded)
        
        # Concatenate all manifolds
        X_ambient = np.vstack(X_ambient_total)
        X_intrinsic = np.vstack(X_intrinsic_total)
        
        # Shuffle the data
        shuffle_idx = np.random.permutation(X_ambient.shape[0])
        X_ambient = X_ambient[shuffle_idx]
        X_intrinsic = X_intrinsic[shuffle_idx]
        
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_fractal_manifold(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate fractal-like manifold with self-similar structure at multiple scales"""
        # Start with base fractal structure (Cantor-like in multiple dimensions)
        X_intrinsic = np.zeros((n_samples, intrinsic_dim))
        
        # Generate hierarchical structure
        for dim in range(intrinsic_dim):
            # Create fractal pattern with multiple scales
            x = np.random.uniform(-1, 1, n_samples)
            
            # Apply fractal transformation at multiple scales
            for scale in range(1, 6):  # 5 different scales
                period = 2**scale
                amplitude = 0.8**(scale-1)
                x += amplitude * np.sin(x * period * np.pi) / period
                x += amplitude * np.cos(x * period * np.pi * 1.618) / period  # Golden ratio for complexity
            
            X_intrinsic[:, dim] = x
        
        # Add cross-dimensional fractal coupling
        for i in range(intrinsic_dim):
            for j in range(i+1, min(i+4, intrinsic_dim)):  # Couple with next few dimensions
                coupling_strength = 0.3 / (j - i)
                X_intrinsic[:, i] += coupling_strength * np.sin(X_intrinsic[:, j] * np.pi * (j-i+1))
        
        # Embedding with fractal-inspired nonlinearities
        embedding_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.3
        
        X_ambient = np.zeros((n_samples, ambient_dim))
        for i in range(ambient_dim):
            linear_comb = X_intrinsic @ embedding_matrix[:, i]
            
            # Apply multi-scale nonlinear transformations
            scale_factor = (i % 8) + 1
            if i % 8 == 0:
                X_ambient[:, i] = np.sin(linear_comb * scale_factor) + 0.5 * np.sin(linear_comb * scale_factor * 3)
            elif i % 8 == 1:
                X_ambient[:, i] = np.tanh(linear_comb * scale_factor) * np.cos(linear_comb * scale_factor * 2)
            elif i % 8 == 2:
                X_ambient[:, i] = linear_comb * np.exp(-linear_comb**2 * scale_factor)
            elif i % 8 == 3:
                X_ambient[:, i] = np.sign(linear_comb) * (np.abs(linear_comb)**0.7) * np.sin(scale_factor * np.pi * linear_comb)
            elif i % 8 == 4:
                X_ambient[:, i] = np.arctan(linear_comb * scale_factor * 2) * np.cosh(linear_comb / scale_factor)
            elif i % 8 == 5:
                X_ambient[:, i] = (linear_comb / (1 + (scale_factor * linear_comb)**2)) * np.sin(scale_factor * linear_comb)
            elif i % 8 == 6:
                X_ambient[:, i] = np.sinh(linear_comb / scale_factor) * np.cos(linear_comb * scale_factor)
            else:
                X_ambient[:, i] = np.cos(linear_comb * scale_factor) * np.exp(-np.abs(linear_comb) / scale_factor)
        
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_attention_like_manifold(n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1):
        """Generate manifold resembling attention mechanism outputs with sparse, head-like structure"""
        n_heads = intrinsic_dim // 5  # Simulate attention heads
        head_dim = intrinsic_dim // n_heads
        
        X_intrinsic = np.zeros((n_samples, intrinsic_dim))
        
        # Generate attention-head-like structure
        for head in range(n_heads):
            start_idx = head * head_dim
            end_idx = min(start_idx + head_dim, intrinsic_dim)
            
            # Each head focuses on different patterns
            focus_strength = np.random.uniform(0.1, 2.0, n_samples)  # Variable attention strength
            
            # Generate query-key-like interactions
            for i in range(start_idx, end_idx):
                base_pattern = np.random.uniform(-1, 1, n_samples)
                
                # Apply attention-like transformation
                if (i - start_idx) % 4 == 0:  # Query-like
                    X_intrinsic[:, i] = base_pattern * focus_strength
                elif (i - start_idx) % 4 == 1:  # Key-like
                    X_intrinsic[:, i] = np.tanh(base_pattern * focus_strength * 2)
                elif (i - start_idx) % 4 == 2:  # Value-like
                    X_intrinsic[:, i] = base_pattern * np.exp(-focus_strength**2/4)
                else:  # Output-like
                    q_idx = start_idx + ((i - start_idx) // 4) * 4
                    k_idx = q_idx + 1
                    v_idx = q_idx + 2
                    if k_idx < intrinsic_dim and v_idx < intrinsic_dim:
                        attention_score = np.tanh(X_intrinsic[:, q_idx] * X_intrinsic[:, k_idx])
                        X_intrinsic[:, i] = attention_score * X_intrinsic[:, v_idx]
                    else:
                        X_intrinsic[:, i] = base_pattern
        
        # Add residual connections and layer norm-like operations
        for i in range(1, intrinsic_dim):
            if i % 7 == 0:  # Residual connections
                X_intrinsic[:, i] += 0.3 * X_intrinsic[:, i-1]
                
        # Layer normalization-like operation per head
        for head in range(n_heads):
            start_idx = head * head_dim
            end_idx = min(start_idx + head_dim, intrinsic_dim)
            head_data = X_intrinsic[:, start_idx:end_idx]
            
            # Normalize
            mean_vals = np.mean(head_data, axis=1, keepdims=True)
            std_vals = np.std(head_data, axis=1, keepdims=True) + 1e-6
            X_intrinsic[:, start_idx:end_idx] = (head_data - mean_vals) / std_vals
        
        # Complex embedding that preserves attention structure
        embedding_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.3
        
        X_ambient = np.zeros((n_samples, ambient_dim))
        
        # Preserve head structure in embedding
        for i in range(ambient_dim):
            head_idx = i % n_heads
            head_start = head_idx * head_dim
            head_end = min(head_start + head_dim, intrinsic_dim)
            
            # Weight more strongly within head
            weights = embedding_matrix[:, i].copy()
            weights[head_start:head_end] *= 2.0  # Stronger intra-head connections
            
            linear_comb = X_intrinsic @ weights
            
            # Apply transformer-like nonlinearities
            if i % 6 == 0:  # GELU approximation
                X_ambient[:, i] = 0.5 * linear_comb * (1 + np.tanh(np.sqrt(2/np.pi) * (linear_comb + 0.044715 * linear_comb**3)))
            elif i % 6 == 1:  # ReLU
                X_ambient[:, i] = np.maximum(0, linear_comb)
            elif i % 6 == 2:  # Swish
                X_ambient[:, i] = linear_comb * (1 / (1 + np.exp(-linear_comb)))
            elif i % 6 == 3:  # Scaled softmax-like
                X_ambient[:, i] = np.exp(linear_comb) / (1 + np.exp(linear_comb))
            elif i % 6 == 4:  # Mish-like
                X_ambient[:, i] = linear_comb * np.tanh(np.log(1 + np.exp(linear_comb)))
            else:  # Leaky ReLU
                X_ambient[:, i] = np.where(linear_comb > 0, linear_comb, 0.01 * linear_comb)
        
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_sparse_coding_manifold(n_samples=2500, intrinsic_dim=35, ambient_dim=250, sparsity=0.7, noise=0.1):
        """Generate manifold with sparse coding structure - simulates sparse neural activations"""
        # Generate sparse dictionary
        dictionary = np.random.randn(ambient_dim, intrinsic_dim * 2) * 0.5
        
        # Generate sparse codes
        X_codes = np.random.randn(n_samples, intrinsic_dim * 2) * 0.8
        
        # Apply sparsity
        sparse_mask = np.random.random((n_samples, intrinsic_dim * 2)) > sparsity
        X_codes *= sparse_mask
        
        # Apply nonlinear activation to sparse codes
        X_codes_activated = np.zeros_like(X_codes)
        for i in range(intrinsic_dim * 2):
            if i % 5 == 0:
                X_codes_activated[:, i] = np.maximum(0, X_codes[:, i])  # ReLU
            elif i % 5 == 1:
                X_codes_activated[:, i] = X_codes[:, i] * (X_codes[:, i] > 0.5)  # Hard threshold
            elif i % 5 == 2:
                X_codes_activated[:, i] = np.tanh(X_codes[:, i] * 2)  # Saturating
            elif i % 5 == 3:
                X_codes_activated[:, i] = X_codes[:, i] * np.exp(-X_codes[:, i]**2/2)  # Gaussian-gated
            else:
                X_codes_activated[:, i] = np.sign(X_codes[:, i]) * np.sqrt(np.abs(X_codes[:, i]))  # Square root
        
        # Reconstruct through dictionary
        X_ambient = X_codes_activated @ dictionary.T
        
        # Add structured noise
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        
        # The intrinsic representation is the activated sparse codes
        X_intrinsic = X_codes_activated[:, :intrinsic_dim]
        
        return X_ambient, X_intrinsic
    
    @staticmethod
    def generate_rnn_like_manifold(n_samples=2500, intrinsic_dim=35, ambient_dim=250, sequence_length=10, noise=0.1):
        """Generate manifold resembling RNN hidden states with temporal dependencies"""
        # Generate temporal sequences
        X_intrinsic = np.zeros((n_samples, intrinsic_dim))
        
        # Pre-generate weight matrices to avoid random seed issues
        forget_weights = np.random.randn(intrinsic_dim, intrinsic_dim) * 0.1
        input_weights = np.random.randn(2*intrinsic_dim//3, intrinsic_dim) * 0.1
        candidate_weights = np.random.randn(2*intrinsic_dim//3, intrinsic_dim) * 0.2
        
        # Initialize with random states
        hidden_state = np.random.randn(intrinsic_dim) * 0.5
        
        # Generate recurrent dynamics
        for sample in range(n_samples):
            # Input at this time step
            input_vec = np.random.randn(intrinsic_dim // 3)
            
            # RNN-like update (simplified LSTM-style)
            for seq_step in range(sequence_length):
                # Forget gate-like
                forget_activation = hidden_state @ forget_weights
                forget_gate = 1 / (1 + np.exp(-np.diag(forget_activation @ forget_weights.T)[:intrinsic_dim]))
                
                # Input gate-like  
                combined_input = np.concatenate([hidden_state[:intrinsic_dim//3], input_vec])
                input_activation = combined_input @ input_weights
                input_gate = 1 / (1 + np.exp(-input_activation[:intrinsic_dim]))
                
                # Update hidden state
                candidate = np.tanh(combined_input @ candidate_weights)
                
                hidden_state = forget_gate * hidden_state + input_gate * candidate[:intrinsic_dim]
                
                # Add nonlinear dynamics
                hidden_state = np.tanh(hidden_state) + 0.1 * np.sin(hidden_state * np.pi)
                
                # Keep values bounded
                hidden_state = np.clip(hidden_state, -3, 3)
            
            X_intrinsic[sample] = hidden_state
            
            # Slight perturbation for next sequence
            hidden_state += np.random.randn(intrinsic_dim) * 0.05
        
        # Complex embedding preserving temporal structure
        embedding_matrix = np.random.randn(intrinsic_dim, ambient_dim) * 0.3
        
        # Add temporal correlation structure to embedding
        for i in range(ambient_dim):
            for j in range(intrinsic_dim):
                # Temporal weighting - recent states matter more
                temporal_weight = np.exp(-abs(i - j) / (ambient_dim / 4))
                embedding_matrix[j, i] *= temporal_weight
        
        X_ambient = np.zeros((n_samples, ambient_dim))
        for i in range(ambient_dim):
            linear_comb = X_intrinsic @ embedding_matrix[:, i]
            
            # Apply RNN-like nonlinearities
            if i % 4 == 0:  # Tanh (classic RNN)
                X_ambient[:, i] = np.tanh(linear_comb)
            elif i % 4 == 1:  # Sigmoid (gates)
                X_ambient[:, i] = 1 / (1 + np.exp(-np.clip(linear_comb, -10, 10)))
            elif i % 4 == 2:  # ReLU (modern RNNs)
                X_ambient[:, i] = np.maximum(0, linear_comb)
            else:  # LSTM-like complex function
                X_ambient[:, i] = linear_comb * (1 / (1 + np.exp(-np.clip(linear_comb, -10, 10))))
        
        X_ambient += np.random.randn(n_samples, ambient_dim) * noise
        return X_ambient, X_intrinsic

class NonlinearityMetrics:
    """Calculate various metrics to assess nonlinearity"""
    
    @staticmethod
    def local_intrinsic_dimensionality_mle(X, k=10):
        """Calculate Local Intrinsic Dimensionality using MLE estimator"""
        n_samples = X.shape[0]
        nbrs = NearestNeighbors(n_neighbors=k+1).fit(X)
        distances, indices = nbrs.kneighbors(X)
        
        # Remove self-distance (first column)
        distances = distances[:, 1:]
        
        lids = []
        for i in range(n_samples):
            dist = distances[i]
            r_max = dist[-1]  # k-th nearest neighbor distance
            
            if r_max > 0:
                # MLE formula for LID
                log_ratios = np.log(r_max / dist[:-1])  # Remove the last distance
                lid = 1 / np.mean(log_ratios) if len(log_ratios) > 0 else 0
            else:
                lid = 0
            
            lids.append(lid)
        
        return np.array(lids)
    
    @staticmethod
    def pca_explained_variance(X, n_components=50):
        """Calculate PCA explained variance ratios"""
        pca = PCA(n_components=min(n_components, X.shape[1], X.shape[0]))
        pca.fit(X)
        return pca.explained_variance_ratio_, np.cumsum(pca.explained_variance_ratio_)
    
    @staticmethod
    def isomap_residual_variance(X, n_neighbors=10, n_components=10):
        """Calculate Isomap residual variance"""
        try:
            isomap = Isomap(n_neighbors=n_neighbors, n_components=n_components)
            X_isomap = isomap.fit_transform(X)
            
            # Calculate geodesic distances
            geodesic_dist = isomap.dist_matrix_
            
            # Calculate Euclidean distances in embedding space
            euclidean_dist = pairwise_distances(X_isomap)
            
            # Flatten upper triangular parts
            mask = np.triu_indices_from(geodesic_dist, k=1)
            geo_flat = geodesic_dist[mask]
            euc_flat = euclidean_dist[mask]
            
            # Remove infinite values
            valid_mask = np.isfinite(geo_flat) & np.isfinite(euc_flat)
            geo_flat = geo_flat[valid_mask]
            euc_flat = euc_flat[valid_mask]
            
            # Calculate residual variance
            correlation = np.corrcoef(geo_flat, euc_flat)[0, 1] if len(geo_flat) > 1 else 0
            residual_variance = 1 - correlation**2
            
            return residual_variance, correlation
        except:
            return np.nan, np.nan
    
    @staticmethod
    def geodesic_euclidean_correlation(X, n_neighbors=15):
        """Calculate correlation between geodesic and Euclidean distances"""
        try:
            n_samples = min(X.shape[0], 1000)  # Subsample for efficiency
            indices = np.random.choice(X.shape[0], n_samples, replace=False)
            X_sub = X[indices]
            
            isomap = Isomap(n_neighbors=n_neighbors, n_components=2)
            isomap.fit(X_sub)
            
            geodesic_dist = isomap.dist_matrix_
            euclidean_dist = pairwise_distances(X_sub)
            
            # Flatten upper triangular parts
            mask = np.triu_indices_from(geodesic_dist, k=1)
            geo_flat = geodesic_dist[mask]
            euc_flat = euclidean_dist[mask]
            
            # Remove infinite values
            valid_mask = np.isfinite(geo_flat) & np.isfinite(euc_flat) & (geo_flat > 0) & (euc_flat > 0)
            geo_flat = geo_flat[valid_mask]
            euc_flat = euc_flat[valid_mask]
            
            if len(geo_flat) > 10:
                pearson_corr = pearsonr(geo_flat, euc_flat)[0]
                spearman_corr = spearmanr(geo_flat, euc_flat)[0]
                return pearson_corr, spearman_corr
            else:
                return np.nan, np.nan
        except:
            return np.nan, np.nan

class AutoencoderAnalysis:
    """Autoencoder for nonlinear dimensionality reduction and reconstruction error analysis"""
    
    def __init__(self, input_dim, encoding_dim, hidden_dims=[128, 64]):
        self.input_dim = input_dim
        self.encoding_dim = encoding_dim
        self.hidden_dims = hidden_dims
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.model = self._build_model()
        
    def _build_model(self):
        layers = []
        dims = [self.input_dim] + self.hidden_dims + [self.encoding_dim]
        
        # Encoder
        for i in range(len(dims) - 1):
            layers.append(nn.Linear(dims[i], dims[i+1]))
            if i < len(dims) - 2:
                layers.append(nn.ReLU())
        
        # Decoder
        dims_rev = dims[::-1]
        for i in range(len(dims_rev) - 1):
            layers.append(nn.Linear(dims_rev[i], dims_rev[i+1]))
            if i < len(dims_rev) - 2:
                layers.append(nn.ReLU())
        
        return nn.Sequential(*layers).to(self.device)
    
    def train(self, X, epochs=100, batch_size=64, lr=0.001):
        X_tensor = torch.FloatTensor(X).to(self.device)
        dataset = TensorDataset(X_tensor)
        dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
        
        optimizer = optim.Adam(self.model.parameters(), lr=lr)
        criterion = nn.MSELoss()
        
        losses = []
        self.model.train()
        for epoch in range(epochs):
            epoch_loss = 0
            for batch in dataloader:
                data = batch[0]
                optimizer.zero_grad()
                reconstructed = self.model(data)
                loss = criterion(reconstructed, data)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
            losses.append(epoch_loss / len(dataloader))
        
        return losses
    
    def get_reconstruction_error(self, X):
        X_tensor = torch.FloatTensor(X).to(self.device)
        self.model.eval()
        with torch.no_grad():
            reconstructed = self.model(X_tensor)
            mse = torch.mean((X_tensor - reconstructed) ** 2, dim=1)
        return mse.cpu().numpy()

def comprehensive_analysis(X, X_intrinsic, dataset_name, k_lid=15):
    """Perform comprehensive nonlinearity analysis"""
    print(f"\nAnalyzing {dataset_name}...")
    
    # Standardize data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    results = {'dataset': dataset_name}
    
    # 1. Local Intrinsic Dimensionality
    print("  Computing Local Intrinsic Dimensionality...")
    lid_values = NonlinearityMetrics.local_intrinsic_dimensionality_mle(X_scaled, k=k_lid)
    results['lid_mean'] = np.mean(lid_values)
    results['lid_std'] = np.std(lid_values)
    results['lid_median'] = np.median(lid_values)
    
    # 2. PCA Analysis
    print("  Computing PCA explained variance...")
    pca_var_ratio, pca_cumsum = NonlinearityMetrics.pca_explained_variance(X_scaled, n_components=50)
    results['pca_first_10_var'] = np.sum(pca_var_ratio[:10])
    results['pca_first_20_var'] = np.sum(pca_var_ratio[:20])
    results['pca_first_30_var'] = np.sum(pca_var_ratio[:30])
    results['pca_90_percent_dims'] = np.argmax(pca_cumsum >= 0.9) + 1
    
    # 3. Geodesic vs Euclidean Distance Correlation
    print("  Computing geodesic-Euclidean correlations...")
    pearson_corr, spearman_corr = NonlinearityMetrics.geodesic_euclidean_correlation(X_scaled)
    results['geodesic_euclidean_pearson'] = pearson_corr
    results['geodesic_euclidean_spearman'] = spearman_corr
    
    # 4. Isomap Residual Variance
    print("  Computing Isomap residual variance...")
    residual_var, iso_corr = NonlinearityMetrics.isomap_residual_variance(X_scaled, n_neighbors=12, n_components=15)
    results['isomap_residual_variance'] = residual_var
    results['isomap_correlation'] = iso_corr
    
    # 5. Kernel PCA comparison
    print("  Computing Kernel PCA analysis...")
    try:
        # RBF Kernel PCA
        kpca_rbf = KernelPCA(n_components=20, kernel='rbf', gamma=1.0/X_scaled.shape[1])
        X_kpca_rbf = kpca_rbf.fit_transform(X_scaled)
        
        # Polynomial Kernel PCA
        kpca_poly = KernelPCA(n_components=20, kernel='poly', degree=3)
        X_kpca_poly = kpca_poly.fit_transform(X_scaled)
        
        results['kpca_rbf_available'] = True
        results['kpca_poly_available'] = True
    except:
        results['kpca_rbf_available'] = False
        results['kpca_poly_available'] = False
    
    # 6. Autoencoder Analysis
    print("  Training autoencoder...")
    try:
        autoencoder = AutoencoderAnalysis(X_scaled.shape[1], encoding_dim=30, hidden_dims=[150, 100, 50])
        losses = autoencoder.train(X_scaled, epochs=50, batch_size=128)
        recon_errors = autoencoder.get_reconstruction_error(X_scaled)
        
        results['autoencoder_final_loss'] = losses[-1]
        results['autoencoder_mean_error'] = np.mean(recon_errors)
        results['autoencoder_std_error'] = np.std(recon_errors)
    except Exception as e:
        print(f"    Autoencoder failed: {e}")
        results['autoencoder_final_loss'] = np.nan
        results['autoencoder_mean_error'] = np.nan
        results['autoencoder_std_error'] = np.nan
    
    # 7. Additional metrics
    results['ambient_dim'] = X.shape[1]
    results['n_samples'] = X.shape[0]
    results['true_intrinsic_dim'] = X_intrinsic.shape[1] if X_intrinsic is not None else np.nan
    
    # Effective rank (numerical rank)
    _, s, _ = np.linalg.svd(X_scaled, full_matrices=False)
    normalized_s = s / s[0]
    results['effective_rank'] = np.sum(normalized_s > 0.01)
    
    return results

def create_comparison_plots(results_df):
    """Create comprehensive comparison plots"""
    plt.figure(figsize=(20, 15))
    
    # Plot 1: LID comparison
    plt.subplot(3, 4, 1)
    plt.bar(results_df['dataset'], results_df['lid_mean'])
    plt.title('Local Intrinsic Dimensionality (Mean)')
    plt.ylabel('LID')
    plt.xticks(rotation=45)
    
    # Plot 2: PCA Explained Variance (first 10 components)
    plt.subplot(3, 4, 2)
    plt.bar(results_df['dataset'], results_df['pca_first_10_var'])
    plt.title('PCA Explained Variance (First 10 Dims)')
    plt.ylabel('Explained Variance Ratio')
    plt.xticks(rotation=45)
    
    # Plot 3: Geodesic-Euclidean Correlation
    plt.subplot(3, 4, 3)
    plt.bar(results_df['dataset'], results_df['geodesic_euclidean_pearson'])
    plt.title('Geodesic-Euclidean Correlation')
    plt.ylabel('Pearson Correlation')
    plt.xticks(rotation=45)
    
    # Plot 4: Isomap Residual Variance
    plt.subplot(3, 4, 4)
    plt.bar(results_df['dataset'], results_df['isomap_residual_variance'])
    plt.title('Isomap Residual Variance')
    plt.ylabel('Residual Variance')
    plt.xticks(rotation=45)
    
    # Plot 5: Autoencoder Reconstruction Error
    plt.subplot(3, 4, 5)
    plt.bar(results_df['dataset'], results_df['autoencoder_mean_error'])
    plt.title('Autoencoder Reconstruction Error')
    plt.ylabel('Mean Squared Error')
    plt.xticks(rotation=45)
    
    # Plot 6: Effective Rank
    plt.subplot(3, 4, 6)
    plt.bar(results_df['dataset'], results_df['effective_rank'])
    plt.title('Effective Rank')
    plt.ylabel('Effective Rank')
    plt.xticks(rotation=45)
    
    # Plot 7: PCA cumulative variance comparison
    plt.subplot(3, 4, 7)
    dims = [10, 20, 30]
    width = 0.25
    x = np.arange(len(results_df))
    
    plt.bar(x - width, results_df['pca_first_10_var'], width, label='First 10', alpha=0.8)
    plt.bar(x, results_df['pca_first_20_var'], width, label='First 20', alpha=0.8)
    plt.bar(x + width, results_df['pca_first_30_var'], width, label='First 30', alpha=0.8)
    
    plt.title('PCA Explained Variance by Dimensions')
    plt.ylabel('Cumulative Explained Variance')
    plt.xticks(x, results_df['dataset'], rotation=45)
    plt.legend()
    
    # Plot 8: Nonlinearity Score (composite)
    plt.subplot(3, 4, 8)
    # Create composite nonlinearity score
    nonlinearity_score = (
        (1 - results_df['geodesic_euclidean_pearson'].fillna(0)) * 0.3 +
        results_df['isomap_residual_variance'].fillna(0) * 0.3 +
        (1 - results_df['pca_first_10_var']) * 0.2 +
        (results_df['lid_mean'] / results_df['lid_mean'].max()) * 0.2
    )
    plt.bar(results_df['dataset'], nonlinearity_score)
    plt.title('Composite Nonlinearity Score')
    plt.ylabel('Nonlinearity Score')
    plt.xticks(rotation=45)
    
    # Plot 9-12: Distribution plots
    for i, metric in enumerate(['lid_mean', 'pca_first_10_var', 'geodesic_euclidean_pearson', 'isomap_residual_variance']):
        plt.subplot(3, 4, 9 + i)
        plt.scatter(range(len(results_df)), results_df[metric], alpha=0.7)
        for j, dataset in enumerate(results_df['dataset']):
            plt.annotate(dataset, (j, results_df[metric].iloc[j]), rotation=90, fontsize=8)
        plt.title(f'{metric} Distribution')
        plt.ylabel(metric.replace('_', ' ').title())
    
    plt.tight_layout()
    plt.show()
    
    # Create correlation matrix
    plt.figure(figsize=(12, 10))
    numeric_cols = results_df.select_dtypes(include=[np.number]).columns
    corr_matrix = results_df[numeric_cols].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
                square=True, linewidths=0.5)
    plt.title('Correlation Matrix of Nonlinearity Metrics')
    plt.tight_layout()
    plt.show()

def main():
    """Main analysis pipeline"""
    print("Generating diverse synthetic datasets...")
    
    # Generate datasets
    datasets = {}
    
    # 1. Linear dataset (baseline)
    print("Generating linear dataset...")
    X_linear, X_linear_intrinsic = DatasetGenerator.generate_linear_dataset(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250
    )
    datasets['Linear'] = (X_linear, X_linear_intrinsic)
    
    # 2. Swiss roll (moderate nonlinearity)
    print("Generating Swiss roll manifold...")
    X_swiss, X_swiss_intrinsic = DatasetGenerator.generate_swiss_roll_high_dim(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.05
    )
    datasets['Swiss Roll'] = (X_swiss, X_swiss_intrinsic)
    
    # 3. Sphere manifold (high nonlinearity)
    print("Generating sphere manifold...")
    X_sphere, X_sphere_intrinsic = DatasetGenerator.generate_sphere_manifold(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.08
    )
    datasets['Sphere Manifold'] = (X_sphere, X_sphere_intrinsic)
    
    # 4. Hypercube corners (discrete/clustered structure)
    print("Generating hypercube corners dataset...")
    X_hypercube, X_hypercube_intrinsic = DatasetGenerator.generate_hypercube_corners(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1
    )
    datasets['Hypercube Corners'] = (X_hypercube, X_hypercube_intrinsic)
    
    # 5. Torus manifold (periodic structure)
    print("Generating torus manifold...")
    X_torus, X_torus_intrinsic = DatasetGenerator.generate_torus_manifold(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1
    )
    datasets['Torus Manifold'] = (X_torus, X_torus_intrinsic)
    
    # 6. Mixture of manifolds (heterogeneous)
    print("Generating mixture of manifolds...")
    X_mixture, X_mixture_intrinsic = DatasetGenerator.generate_mixture_of_manifolds(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1
    )
    datasets['Mixture Manifolds'] = (X_mixture, X_mixture_intrinsic)
    
    # 7. Fractal manifold (multi-scale structure)
    print("Generating fractal manifold...")
    X_fractal, X_fractal_intrinsic = DatasetGenerator.generate_fractal_manifold(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.12
    )
    datasets['Fractal Manifold'] = (X_fractal, X_fractal_intrinsic)
    
    # 8. Attention-like manifold (transformer-style)
    print("Generating attention-like manifold...")
    X_attention, X_attention_intrinsic = DatasetGenerator.generate_attention_like_manifold(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, noise=0.1
    )
    datasets['Attention-like'] = (X_attention, X_attention_intrinsic)
    
    # 9. Sparse coding manifold
    print("Generating sparse coding manifold...")
    X_sparse, X_sparse_intrinsic = DatasetGenerator.generate_sparse_coding_manifold(
        n_samples=2500, intrinsic_dim=35, ambient_dim=250, sparsity=0.7, noise=0.1
    )
    datasets['Sparse Coding'] = (X_sparse, X_sparse_intrinsic)
    
    
    
    # Analyze all datasets
    results = []
    for name, (X, X_intrinsic) in datasets.items():
        result = comprehensive_analysis(X, X_intrinsic, name)
        results.append(result)
    
    # Convert to DataFrame
    results_df = pd.DataFrame(results)
    
    # Display results
    print("\n" + "="*100)
    print("COMPREHENSIVE NONLINEARITY ANALYSIS RESULTS - DIVERSE DATASETS")
    print("="*100)
    
    # Key metrics table
    key_metrics = [
        'dataset', 'lid_mean', 'pca_first_10_var', 'geodesic_euclidean_pearson', 
        'isomap_residual_variance', 'autoencoder_mean_error', 'effective_rank'
    ]
    
    print("\nKey Nonlinearity Metrics:")
    print(results_df[key_metrics].round(4))
    
    # Extended metrics table
    extended_metrics = [
        'dataset', 'lid_median', 'pca_first_20_var', 'pca_first_30_var', 
        'geodesic_euclidean_spearman', 'isomap_correlation', 'autoencoder_final_loss'
    ]
    
    print("\nExtended Metrics:")
    print(results_df[extended_metrics].round(4))
    
    # Interpretation guide
    print("\n" + "="*100)
    print("INTERPRETATION GUIDE FOR DIVERSE DATASETS")
    print("="*100)
    print("""
    Dataset Characteristics:
    
    Linear: Random projection baseline - should show high PCA variance, low LID
    Swiss Roll: Moderate nonlinearity with smooth manifold structure  
    Sphere Manifold: High nonlinearity with global curvature constraints
    Hypercube Corners: Discrete clusters simulating sparse activations
    Torus Manifold: Periodic/oscillatory structure with multiple frequencies
    Mixture Manifolds: Heterogeneous structure combining different manifold types
    Fractal Manifold: Multi-scale self-similar structure
    Attention-like: Transformer-style with head structure and normalization
    Sparse Coding: Dictionary learning structure with sparsity constraints
    RNN-like: Temporal dependencies with recurrent dynamics
    
    Nonlinearity Indicators:
    
    Local Intrinsic Dimensionality (LID):
    - Linear: ~35 (matches true intrinsic dim)
    - Moderate nonlinear: 25-45 
    - High nonlinear: >50 or highly variable
    - Structured (attention, sparse): May show bimodal distribution
    
    PCA Explained Variance (First 10 dims):
    - Linear: >0.8
    - Moderate nonlinear: 0.3-0.7
    - High nonlinear: <0.3
    - Structured: Depends on structure type
    
    Geodesic-Euclidean Correlation:
    - Linear: >0.95
    - Moderate nonlinear: 0.5-0.9  
    - High nonlinear: <0.5
    - Discrete/clustered: May be low due to disconnected regions
    
    Expected Nonlinearity Ranking (Low to High):
    1. Linear (baseline)
    2. Swiss Roll (smooth manifold)
    3. Torus Manifold (periodic but smooth)
    4. RNN-like (temporal smoothing)
    5. Attention-like (structured but normalized)
    6. Sphere Manifold (global constraints)
    7. Sparse Coding (dictionary constraints)
    8. Hypercube Corners (discrete jumps)
    9. Mixture Manifolds (heterogeneous)
    10. Fractal Manifold (multi-scale complexity)
    """)
    
    # Nonlinearity ranking
    print("\n" + "="*60)
    print("NONLINEARITY RANKING (Low to High)")
    print("="*60)
    
    # Create enhanced composite score
    # Normalize metrics for fair comparison
    results_df_norm = results_df.copy()
    
    # Handle missing values and normalize
    geo_euc_corr = results_df['geodesic_euclidean_pearson'].fillna(results_df['geodesic_euclidean_pearson'].mean())
    iso_residual = results_df['isomap_residual_variance'].fillna(results_df['isomap_residual_variance'].mean())
    autoenc_error = results_df['autoencoder_mean_error'].fillna(results_df['autoencoder_mean_error'].mean())
    
    # Compute composite nonlinearity score
    nonlinearity_scores = (
        (1 - geo_euc_corr) * 0.25 +  # Lower correlation = higher nonlinearity
        iso_residual * 0.25 +        # Higher residual = higher nonlinearity  
        (1 - results_df['pca_first_10_var']) * 0.25 +  # Lower PCA explained = higher nonlinearity
        (autoenc_error / autoenc_error.max()) * 0.15 +  # Higher AE error = more complex
        ((results_df['lid_mean'] - results_df['true_intrinsic_dim'].fillna(35))**2).fillna(0) / 100 * 0.1  # LID deviation from true dim
    )
    
    ranking_df = pd.DataFrame({
        'Dataset': results_df['dataset'],
        'Nonlinearity Score': nonlinearity_scores,
        'PCA First 10': results_df['pca_first_10_var'],
        'Geo-Euc Corr': geo_euc_corr,
        'Isomap Residual': iso_residual,
        'LID Mean': results_df['lid_mean'],
        'AE Error': autoenc_error
    }).sort_values('Nonlinearity Score')
    
    print(ranking_df.round(4))
    
    # Identify datasets meeting the specific criteria
    print("\n" + "="*80)
    print("DATASETS MEETING HIGH COMPLEXITY CRITERIA")
    print("="*80)
    
    criteria_met = results_df[
        (results_df['geodesic_euclidean_pearson'] < 0.4) & 
        (results_df['pca_first_10_var'] < 0.3)
    ]
    
    if len(criteria_met) > 0:
        print("Datasets with low Geodesic-Euclidean Correlation (<0.4) AND low PCA explained variance (<0.3):")
        print(criteria_met[['dataset', 'geodesic_euclidean_pearson', 'pca_first_10_var', 'lid_mean', 'isomap_residual_variance']].round(4))
    else:
        print("No datasets fully meet both criteria. Relaxing constraints...")
        relaxed_criteria = results_df[
            (results_df['geodesic_euclidean_pearson'] < 0.6) & 
            (results_df['pca_first_10_var'] < 0.5)
        ]
        print("Datasets with moderate complexity:")
        print(relaxed_criteria[['dataset', 'geodesic_euclidean_pearson', 'pca_first_10_var', 'lid_mean']].round(4))
    
    # Create visualizations
    create_comparison_plots(results_df)
    
    # Additional analysis: Correlation between different nonlinearity measures
    print("\n" + "="*60)
    print("CORRELATION BETWEEN NONLINEARITY MEASURES")
    print("="*60)
    
    nonlinearity_metrics = ['lid_mean', 'pca_first_10_var', 'geodesic_euclidean_pearson', 
                           'isomap_residual_variance', 'autoencoder_mean_error']
    
    available_metrics = [m for m in nonlinearity_metrics if m in results_df.columns and not results_df[m].isna().all()]
    
    if len(available_metrics) >= 2:
        corr_matrix = results_df[available_metrics].corr()
        print(corr_matrix.round(3))
        
        # Insights
        print("\nKey Insights:")
        if 'pca_first_10_var' in available_metrics and 'geodesic_euclidean_pearson' in available_metrics:
            pca_geo_corr = corr_matrix.loc['pca_first_10_var', 'geodesic_euclidean_pearson']
            print(f"- PCA vs Geodesic-Euclidean correlation: {pca_geo_corr:.3f}")
            if pca_geo_corr > 0.7:
                print("  Strong positive correlation suggests both capture similar linear structure")
            elif pca_geo_corr < 0.3:
                print("  Weak correlation suggests they capture different aspects of nonlinearity")
        
        if 'lid_mean' in available_metrics and 'isomap_residual_variance' in available_metrics:
            lid_iso_corr = corr_matrix.loc['lid_mean', 'isomap_residual_variance']
            print(f"- LID vs Isomap residual correlation: {lid_iso_corr:.3f}")
            if lid_iso_corr > 0.5:
                print("  Positive correlation suggests both capture manifold complexity")
    
    return results_df

if __name__ == "__main__":
    results = main()
```
