```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from data_generator import generate_2d_clustered_data
from utils import plot_pca_data, plot_kernel_pca_comparison

```

```python
X, y, original_centers = generate_2d_clustered_data(
        n_samples=350, 
        n_clusters=7, 
        d=100, 
        circle_radius=8, 
        cluster_std=0.2,
        random_state=42
)
    
print(f"Generated dataset shape: {X.shape}")
print(f"Number of clusters: {len(np.unique(y))}")
print(f"Cluster sizes: {np.bincount(y)}")
    
    
```

```python
X_pca, pca_model = plot_pca_data(X, y, "Generated Clustered Data (High-Dimensional)")
```

```python
kpca_results = plot_kernel_pca_comparison(X, y, 
                                            kernels=[ 'rbf', 'cosine', 'sigmoid', 'poly'],
                                            title="Kernel PCA with Different Kernels")
```

```python
from curved_steering import CurvedSteering
```

```python
steering = CurvedSteering(X, y, method='pca', kernel = 'sigmoid')  # or 'kernel_pca'

# Visualize the steering process
steered_points, steering_info_list = steering.visualize_steering(80, step_size=0.3, direction=1, n_steps=50)
```

```python
# Create any random point 
my_random_point = np.random.normal(0, 1, size=100) 

# See how the random point gets projected and steered
steered_points, _ = steering.visualize_arbitrary_point_steering(
    point=my_random_point,
    step_size=0.2,
    direction=1,
    n_steps=20
)
```
